## **PREPROCESSING**

```python
df = main_df.copy()
```

### *Data Type Transform*

```python
#change Dt_Customer dtype from object to datetime 64[ns]
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
import datetime as dt
from datetime import date
```

### *Feature Extraction*

```python
#create total accepted campaign
df['Total_Acc_Cmp'] = df['AcceptedCmp1']+df['AcceptedCmp2']+df['AcceptedCmp3']+df['AcceptedCmp4']+df['AcceptedCmp5']

#create total purchases feature
df['Total_Purchases'] = df['NumDealsPurchases']+df['NumWebPurchases']+df['NumCatalogPurchases']+df['NumStorePurchases']

#create conversion rate feature
def safe_div(x,y):
    if y == 0:
        return 0
    return x / y
df['cvr'] = df.apply(lambda x: safe_div(x['Total_Purchases'],x['NumWebVisitsMonth']), axis=1)

#create age feature
df['Age'] = 2022 - df['Year_Birth']

#create age group
age_list=[]
for i in df['Age']:
    if i >= 0 and i <= 1:
        group = 'Infant'
    elif i >= 2 and i <= 4:
        group = 'Toddler'
    elif i >= 5 and i <= 12:
        group = 'Child'
    elif i >= 13 and i <= 19:
        group = 'Teen'
    elif i >= 20 and i <= 39:
        group = 'Adult'
    elif i >= 40 and i <= 59:
        group = 'Middle Aged'
    else:
        group = 'Senior Citizen'
    age_list.append(group)
df['Age_Group'] = age_list

#create total spend feature
df['Total_Spent'] = df['MntCoke']+df['MntFishProducts']+df['MntFruits']+df['MntMeatProducts']+df['MntSweetProducts']+df['MntGoldProds']

#create ammount of children feature
df['NumChildren'] = df['Kidhome']+df['Teenhome']

#create total of days joined
df['Dt_Collected'] = date.today()
df['Dt_Collected'] = df['Dt_Collected'].astype('datetime64[ns]')
df['Dt_Days_Customer'] = df['Dt_Collected'] - df['Dt_Customer']
df['Dt_Days_Customer'] = df['Dt_Days_Customer'].dt.days
```

## *Feature Encoding*

```python
from sklearn.preprocessing import LabelEncoder
lenc = LabelEncoder()
lenc.fit(df['Marital_Status'])
df['Marital_Status'] = lenc.transform(df['Marital_Status'])
from sklearn.preprocessing import OrdinalEncoder
edu = ['D3', 'S1', 'S2', 'S3', 'SMA']
ore = OrdinalEncoder(categories=([edu]))
ore.fit(df[['Education']])
df['Education'] = ore.transform(df[['Education']])
```

### Handling Missing Value

```python
df.isna().sum()
```

```python
df['Income'].fillna(df['Income'].mean(), inplace=True)
```

### *Drop Unnecesary Feature*

```python
df.drop(['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5',
                'NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases',
                'MntCoke','MntFishProducts','MntFruits','MntMeatProducts','MntSweetProducts', 'MntGoldProds',
                'Kidhome','Teenhome', 'Response','Z_CostContact','Z_Revenue','Dt_Collected'], inplace=True, axis=1)
```

### *Handling Outliers*

```python
cols = df_rfm.columns
plt.figure(figsize= (10,15))
for i in range(len(cols)):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y = df_rfm[cols[i]], orient='v')
    plt.tight_layout()
```

```python
for col in cols:
    high_cut = df_rfm[col].quantile(q=0.99)
    low_cut= df_rfm[col].quantile(q=0.01)
    df_rfm.loc[df_rfm[col]>high_cut,col]=high_cut
    df_rfm.loc[df_rfm[col]<low_cut,col]=low_cut
```

```python
cols = df_rfm.columns
plt.figure(figsize= (10,15))
for i in range(len(cols)):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y = df_rfm[cols[i]], orient='v')
    plt.tight_layout()
```
